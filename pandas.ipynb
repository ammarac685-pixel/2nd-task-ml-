```python
import pandas as pd

data = {
    "product": ["A", "B", "C", "A", "B", "C", "D", "E", "F", "A"],
    "region": ["North", "South", "North", "East", "North", "South", "North", "East", "North", "South"],
    "date": pd.date_range(start="2024-01-01", periods=10),
    "quantity": [10, 5, 8, 6, 7, 3, 9, 4, 2, 11],
    "price": [100, 200, 150, 120, 180, 160, 140, 130, 170, 110]
}

df = pd.DataFrame(data)

print("Full Data:\n", df)
print("\nShape:", df.shape)
print("\nColumns:", df.columns)

filtered = df[df["region"] == "North"]
print("\nFiltered Data (North):\n", filtered)

filtered["revenue"] = filtered["quantity"] * filtered["price"]
print("\nWith Revenue:\n", filtered)

grouped = filtered.groupby("product")["revenue"].sum().reset_index()
print("\nTotal Revenue per Product:\n", grouped)

top5 = grouped.sort_values(by="revenue", ascending=False).head(5)
print("\nTop 5 Products:\n", top5)

meta = pd.DataFrame({
    "product": ["A", "B", "C", "D", "E", "F"],
    "category": ["Electronics", "Clothing", "Food", "Electronics", "Food", "Clothing"]
})

final = pd.merge(top5, meta, on="product")
print("\nFinal Merged Data:\n", final)

final.to_csv("final_output.csv", index=False)
```
