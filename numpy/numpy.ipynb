```python
import numpy as np

A = np.random.randint(1, 10, (3, 3))
B = np.random.randint(1, 10, (3, 3))

print("Matrix A:\n", A)
print("Shape of A:", A.shape)

print("\nMatrix B:\n", B)
print("Shape of B:", B.shape)

elementwise = A * B
print("\nElement-wise Multiplication:\n", elementwise)

matrix_mult = np.dot(A, B)
print("\nMatrix Multiplication:\n", matrix_mult)

det_A = np.linalg.det(A)
print("\nDeterminant of A:", det_A)

inv_A = np.linalg.inv(A)
print("\nInverse of A:\n", inv_A)

arr = np.random.randint(1, 20, 5)
print("\nOriginal 1D Array:", arr)

normalized = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
print("Normalized Array:", normalized)

C = np.random.randint(1, 10, (4, 3))
print("\nMatrix C:\n", C)

col_mean = np.mean(C, axis=0)
print("Column Means:", col_mean)

broadcasted = C - col_mean
print("\nAfter Broadcasting (Column Mean Subtracted):\n", broadcasted)
```
